### Preparing the data

```python
# Separating label from features

X=df.drop("Class", axis=1)
y=df["Class"]
X.shape, y.shape
```

```python
# Separating Train and Test set
X_train, X_test, y_train, y_test=train_test_split(X, y, test_size=0.3, stratify=y, random_state=1)
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
# Scaling all data to have a clean distribution

scale=StandardScaler()

scale.fit(X_train['Amount'].values.reshape(-1,1))
X_train['scaled_amount']=scale.transform(X_train['Amount'].values.reshape(-1,1))
X_test['scaled_amount']=scale.transform(X_test['Amount'].values.reshape(-1,1))

scale.fit(X_train['Time'].values.reshape(-1,1))
X_train['scaled_time']=scale.transform(X_train['Time'].values.reshape(-1,1))
X_test['scaled_time']=scale.transform(X_test['Time'].values.reshape(-1,1))

X_train.drop(['Time','Amount'], axis=1, inplace=True)
X_test.drop(['Time','Amount'], axis=1, inplace=True)
```

```python
print("Train", y_train.value_counts()/X_train.shape[0]*100)
print("Test", y_test.value_counts()/X_test.shape[0]*100)
```

```python
df_train_scaled = pd.concat([X_train, y_train], axis=1)

co_mat=df_train_scaled.corr()

plt.figure(figsize=(17,17))
sns.heatmap(co_mat, annot=True, fmt=".1f", cmap='coolwarm_r')
```

## Model Development

```python
# Calculate the number of each class and weight
neg = len(y_train[y_train == 0])
pos = len(y_train[y_train == 1])

ratio = neg / pos
```

```python
XGB=xgb.XGBClassifier(scale_pos_weight=ratio, max_depth=9)
XGB.fit(X_train, y_train)
XGB.score(X_train, y_train), XGB.score(X_test, y_test)
```

```python
y_pred=XGB.predict(X_test)
print(classification_report(y_test, y_pred, labels=XGB.classes_))
```

```python
f, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(18,6))

fpr, tpr, thresholds = roc_curve(y_test, y_pred)
roc_auc = auc(fpr, tpr)
ROC=RocCurveDisplay(fpr=fpr, tpr=tpr, roc_auc=roc_auc, name='XGBoost', pos_label=1)
ROC.plot(ax=ax1)
ax1.set_title("ROC")

PrecisionRecallDisplay.from_estimator(XGB, X_test, y_test, name="XGBoost", ax=ax2)
ax2.set_title("Percision-Recall")

cfx = ConfusionMatrixDisplay.from_predictions(y_test, y_pred, ax=ax3, colorbar=False)
ax3.set_title("Confusion Matrix")

plt.savefig("../XGBFinalResults.pdf", bbox_inches="tight")
```
